```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

def recalculate_clusters(X, centroids, k):
   
    clusters = {}
    for i in range(k):
        clusters[i] = []
    for data in X:
        euc_dist = []
        for j in range(k):
            euc_dist.append(np.linalg.norm(data - centroids[j]))
        clusters[euc_dist.index(min(euc_dist))].append(data)
    print("\nnew clusters : " , clusters)
    return clusters

def recalculate_centroids(centroids, clusters, k):
    for i in range(k):
        centroids[i] = np.average(clusters[i], axis=0)
    return centroids

X = pd.read_csv("~/1nt18is036/points.csv")
print(X)
print(type(X))
k=3
# print(X['x'].values)
centroids = []
clusters = {}
for i in range(k):
    clusters[i] = []
for i in range(k):
    centroids.append([X['x'][i] , X['y'][i]])
# centroids.append([X['x'][4] , X['y'][4]])
# centroids.append([X['x'][5] , X['y'][5]])
# centroids.append([X['x'][6] , X['y'][6]])
print("\ncentroids are : " , centroids)

X = X[['x', 'y']]
dataset = X.astype(float).values.tolist()

points = X.values #returns a numpy array
# plt.scatter(points[:,0],points[:,1], label='True Position')
kmeans = KMeans(n_clusters=3)
kmeans.fit(points)
print("\nk-means cluster center" ,kmeans.cluster_centers_)


print("\n", "centroid" , "  point-x,y" , "   e-distance")

for i in points:
    e_dist = []
    print("\n")
    for j in range(k):
        e_dist.append(np.linalg.norm(i - centroids[j]))
        print( centroids[j] , i , "   ", e_dist[j])
    #print("\n",e_dist)
    clusters[e_dist.index(min(e_dist))].append(i)
    
print("\nclusters assigned : " , clusters)
new_centroids = []


for i in range(k):
        new_centroids.append([X['x'][i] , X['y'][i]])
for i in range(k):
        new_centroids[i] = np.average(clusters[i])


while (np.array_equal(centroids , new_centroids) is False):
   
   
    
  
   
    clusters = recalculate_clusters(points,new_centroids,k)
   
    new_centroids = recalculate_centroids(centroids , clusters, k)    
    centroids = new_centroids
    print("\nnew centroids are",new_centroids)

            
    print("\nold centroids were",centroids)
    
    print("\nk-means cluster center" ,kmeans.cluster_centers_)

    
if(np.array_equal(centroids , new_centroids)):
    print("\nK-means completed") 
    


    


    
```
